```python
dataframe = dataframe[["id","fields/issuetype/name","fields/project/name","fields/status/statusCategory/name","fields/description","fields/summary"]]
```

```python
dataset = dataframe
```

```python
# Text Processing Function
def text_processing(dataset):
  # Removing contents of tags and all for further text processing
  dataset['fields/description'].replace(regex=True,inplace=True, to_replace= r'<.+?>', value=r' ')
  dataset['fields/summary'].replace(regex=True,inplace=True, to_replace= r'<.+?>', value=r' ')

  # Removing the \t charecter from the descriptions field
  dataset['fields/description'] = dataset['fields/description'].str.replace(r'\t', '')

  # Removing links from all for further text processing
  dataset['fields/description'].replace(regex=True,inplace=True, to_replace= r'(https?:\/\/)(\s)*(www\.)?(\s)*((\w|\s)+\.)*([\w\-\s]+\/)*([\w\-]+)((\?)?[\w\s]*=\s*[\w\%&]*)*', value=r' ')
  dataset['fields/summary'].replace(regex=True,inplace=True, to_replace= r'(https?:\/\/)(\s)*(www\.)?(\s)*((\w|\s)+\.)*([\w\-\s]+\/)*([\w\-]+)((\?)?[\w\s]*=\s*[\w\%&]*)*', value=r' ')

  # Remove punctuation
  dataset['fields/description'].replace(regex=True,inplace=True, to_replace= r'[^\w\d\s]', value=r'')
  dataset['fields/summary'].replace(regex=True,inplace=True, to_replace= r'[^\w\d\s]', value=r'')

  # Removing rows with empty columns 
  dataset.dropna(subset=['fields/description','fields/summary'],inplace=True)

  return dataset 

# def remove_punctuation(text):
#   # Removing the punctuations
#   punctuationfree="".join([i for i in text if i not in string.punctuation])
#   return punctuationfree
```

```python
dataset = text_processing(dataset)

# dataset['clean_desc'] = dataset['fields/description'].apply(lambda x:remove_punctuation(x))
# dataset['clean_desc'] = dataset['fields/description'].apply(lambda x:remove_tabs(x))
# dataset['clean_summ'] = dataset['fields/summary'].apply(lambda x:remove_punctuation(x))
# dataset['clean_summ'] = dataset['fields/summary'].apply(lambda x:remove_tabs(x))
```
